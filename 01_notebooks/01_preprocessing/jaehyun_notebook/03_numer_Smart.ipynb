```python
import pandas as pd
import numpy as np
```

```python
# 사분위수 활용 연속형 변수로 변환
# 데이터 불러오기
df = pd.read_csv("../../../00_data/01_interim/numerical.csv")

# 1. 로그 변환
df["log_ssdmart"] = np.log(df["p__c01001"] + 1)

# 2. 5구간 생성 (분위수 기준)
df["Sm_group"] = pd.qcut(
    df["log_mobile_cost"],
    q=5,
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)

# 3. 머신러닝용 숫자 코드 변수
df["mobile_cost_group_code"] = pd.qcut(
    df["log_mobile_cost"],
    q=5,
    labels=False
)

# 4. 고사용자 여부 변수 (상위 40%)
df["mobile_cost_high_user"] = (df["mobile_cost_group_code"] >= 3).astype(int)

# 5. 분포 확인
print(df["mobile_cost_group"].value_counts())

# 6. 구간 기준값 확인
bins = pd.qcut(df["log_mobile_cost"], q=5, retbins=True)

# 로그값을 원래 금액 기준으로 변환
print(np.exp(bins[1]) - 1)
```

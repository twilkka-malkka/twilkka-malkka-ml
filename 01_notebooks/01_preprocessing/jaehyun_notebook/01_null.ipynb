```python
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv("../../../00_data/01_interim/first_filtered_data.csv")

# 1. 데이터 기본 정보
print("데이터 shape")
print(df.shape)

print("\n컬럼 정보")
print(df.info())

# 2. 컬럼별 결측치 개수
missing_count = df.isnull().sum()

# 3. 컬럼별 결측치 비율
missing_ratio = (df.isnull().sum() / len(df)) * 100

# 4. 정리된 테이블 생성
missing_table = pd.DataFrame({
    "missing_count": missing_count,
    "missing_ratio(%)": missing_ratio
})

# 결측치 있는 컬럼만 보기
missing_table = missing_table[missing_table["missing_count"] > 0]
missing_table = missing_table.sort_values(by="missing_ratio(%)", ascending=False)

print("\n결측치 요약")
print(missing_table)
```

```python
import pandas as pd
import numpy as np
```

```python
# 사분위수 활용 연속형 변수로 변환
# 데이터 불러오기
df = pd.read_csv("../../../00_data/01_interim/numerical.csv")

# 1. 로그 변환
df["log_mobile_cost"] = np.log(df["p__c01001"] + 1)

# 2. 5구간 생성 (분위수 기준)
df["mobile_cost_group"] = pd.qcut(
    df["log_mobile_cost"],
    q=5,
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)

# 3. 머신러닝용 숫자 코드 변수
df["mobile_cost_group_code"] = pd.qcut(
    df["log_mobile_cost"],
    q=5,
    labels=False
)

# 4. 고사용자 여부 변수 (상위 40%)
df["mobile_cost_high_user"] = (df["mobile_cost_group_code"] >= 3).astype(int)

# 5. 분포 확인
print(df["mobile_cost_group"].value_counts())

# 6. 구간 기준값 확인
bins = pd.qcut(df["log_mobile_cost"], q=5, retbins=True)

# 로그값을 원래 금액 기준으로 변환
print(np.exp(bins[1]) - 1)
```

```python
from sklearn.preprocessing import StandardScaler

# 1. 로그 변환 (금액 변수)
df["log_mobile_cost"] = np.log(df["p__c01001"] + 1)

# 2. 표준화 (평균 0, 표준편차 1)
scaler = StandardScaler()
df["log_mobile_cost_scaled"] = scaler.fit_transform(df[["log_mobile_cost"]])

# 결과 확인
print(df[["p__c01001", "log_mobile_cost", "log_mobile_cost_scaled"]].head())
```

```python
from sklearn.preprocessing import MinMaxScaler

# 정규화 객체 생성
scaler = MinMaxScaler()

# 정규화 수행 (0~1 범위)
df["mobile_cost_normalized"] = scaler.fit_transform(df[["p__c01001"]])

# 결과 확인
print(df[["p__c01001", "mobile_cost_normalized"]].head())
```
